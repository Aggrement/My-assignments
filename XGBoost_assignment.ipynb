Import liberaries

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import xgboost as xgb
```

Data loading and preparation

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

Load the dataset

```python
df=pd.read_csv("Breast cancer.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

As column "id" contains patient or image ID and it is not a feature; therefore, I drop column "id".

```python
df=df.drop("id", axis=1)
```

```python
df.head()
```

As last column (ie, column "Unnamed: 32") is empty; therefore, I dropped this column.

```python
df=df.drop("Unnamed: 32", axis=1)
df.head()
```

```python
df.info()
```

Perform onehot encoding on Column "diagnosis"

```python
df["diagnosis"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
df["diagnosis"]=encoder.fit_transform(df["diagnosis"])
df["diagnosis"].unique()
```

```python
df.head()
```

Split the data into X and Y

```python
X=df.iloc[:,1:31].values
y=df.iloc[:,:1].values
```

Split the data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.3, random_state=10)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

Hyperparameter Tuning:

```python
model=XGBClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

Define hyperparameters

```python
param_grid={"max_depth":[4, 5, 6],"n_estimators": [500, 600, 700],"learning_rate":[0.01, 0.015]}
```

Perform grid search and find the best parameters

```python
search = GridSearchCV(model, param_grid, cv=5).fit(X_train, y_train)
```

```python
print(search.best_params_)
```

Use values for hyperparameters in model

```python
model=XGBClassifier(learning_rate=search.best_params_["learning_rate"], n_estimators=search.best_params_["n_estimators"], max_depth=search.best_params_["max_depth"])
```

```python
model.fit(X_train, y_train)
```

Obtain predictions from trained model on test data.

```python
y_pred=model.predict(X_test)
```

Check accuracy

```python
accuracy=accuracy_score(y_test, y_pred)*100
print (accuracy)
```
